### Data Visualisations and Evaluation

#### A’s required at least 99 wins in order to make the playoffs in 2002.

```python
# Plotting scatterplot
plt.figure(figsize=(10,8))
moneyball1=moneyball.loc[moneyball['Playoffs']==1]
moneyball0=moneyball.loc[moneyball['Playoffs']==0]
plt.scatter(x=moneyball0.W,y = moneyball0.RS,c="lightgreen")
plt.scatter(x=moneyball1.W,y = moneyball1.RS,c="magenta")
plt.xlabel("Wins")
plt.ylabel("Runs Scored")
plt.title("Estimation of no of wins required for the playoffs")
plt.legend(('not qualified','qualified'))
plt.axvline(x = 99)
```

from the above graph, we can see that only 3 teams with wins>=99 have not been Qualified, so it can be estimated that teams with minimum of 99 wins can make it up to the playoffs.

```python
#plots to find insights regarding the distributions of the numerical variables.

#to plot four subplots
fig,fpos=plt.subplots(2,2)
fpos[0,0].hist(moneyball.RS,bins=50)
fpos[0,0].set_title("Runs scored")
fpos[0,1].hist(moneyball.RA,bins=50)
fpos[0,1].set_title("Runs Allowed")
fpos[1,0].hist(moneyball.OBP,bins=50)
fpos[1,0].set_title("On Base Percentage")
fpos[1,1].hist(moneyball.SLG,bins=50)
fpos[1,1].set_title("Slugging Percentage")
fig.subplots_adjust(hspace=0.2)
fig.set_figheight(10)
fig.set_figwidth(10)
```

from the above graphs, we can see that both the graphs of runs scored and runs allowed are more populated to the left. The 'tail' of runs scored is prominant than that of the runs allowed.

#### The A’s needed to score atleast 814 runs and allow only 645 runs in order to make it to the postseason. That meant that the difference between runs scored and runs allowed needed to be approximately 169(RD).

```python
x = np.array(moneyball.RD)
y = np.array(moneyball.W)
# Deriving slope,intercept values
slope, intercept = np.polyfit(x, y, 1)
abline_values = [slope * i + intercept for i in x]
#Plotting the figure
plt.figure(figsize=(10,8))
plt.scatter(x, y,c="lightblue")
plt.plot(x, abline_values, 'b')
plt.title("Relation b/w RD and W (Slope = %s)" % (slope))
plt.xlabel("Run Difference")
plt.ylabel("Wins")
plt.show()
```

from the above graph, we can see that the relation between RD and W is "linear".

```python
#calculaing the co-relation
print(np.corrcoef(x,y))
```

```python
#plots to represent the relation between OBP, SLG and BA with the runs scored respectively.
y = np.array(moneyball.RS)
#plot RS vs OBP
x = np.array(moneyball.OBP)
slope, intercept = np.polyfit(x,y,1)
albine_values = [slope * i + intercept for i in x]
plt.figure(figsize=(10,8))
plt.scatter(x,y,c="lightpink")
plt.plot(x,albine_values,'b')
plt.title("Relation b/w OBP and RS (Slope = %s)" % (slope))
plt.xlabel("On-base percentage")
plt.ylabel("Runs Scored")
plt.show()

#plot RS vs SLG
x = np.array(moneyball.SLG)
slope, intercept = np.polyfit(x,y,1)
albine_values = [slope * i + intercept for i in x]
plt.figure(figsize=(10,8))
plt.scatter(x,y,c="lightblue")
plt.plot(x,albine_values,'b')
plt.title("Relation b/w SLG and RS (Slope = %s)" % (slope))
plt.xlabel("slugging percentage")
plt.ylabel("Runs Scored")
plt.show()

#plot RS vs BA
x = np.array(moneyball.BA)
slope, intercept = np.polyfit(x,y,1)
albine_values = [slope * i + intercept for i in x]
plt.figure(figsize=(10,8))
plt.scatter(x,y,c="lightgrey")
plt.plot(x,albine_values,'b')
plt.title("Relation b/w BA and RS (Slope = %s)" % (slope))
plt.xlabel("batting-avg percentage")
plt.ylabel("Runs Scored")
plt.show()
```

from the above plots,we can infer that there is a clear linear relationship between each of the variables and runs scored.

### Regression Models 

```python
#using scikit-learn library to build our models.
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

#### To predict runs scored

```python
# Extracting our variables from the dataframe.
x = moneyball[['OBP','SLG','BA']].values
y = moneyball[['RS']].values
```

```python
# Calling our model object.
RS_model = LinearRegression()
# Fitting the model.
RS_model.fit(x,y)
```

```python
# Printing model intercept and coefficients.
print(RS_model.intercept_)
print(RS_model.coef_)
```

      
Thus, our model is of the following form:   
RS = -788.46+2917.42×(OBP)+1637.93×(SLG)-368.97×(BA)      

here we observe that BA is negative, and it represents "multi-collinearity" phenomena which causes discrepancies in multiple linear regression models.    
In order to avoid multi-collinearity, we must either add an interaction variable or remove the variable causing the discrepancy.

```python
#removing BA
x = moneyball[['OBP','SLG']].values
y = moneyball[['RS']].values
```

```python
RS_model = LinearRegression()
RS_model.fit(x,y)
```

```python
# Printing model intercept and coefficients.
print(RS_model.intercept_)
print(RS_model.coef_)
```

the model is of the form:  
RS = -804.63+2737.77×(OBP)+1584.91×(SLG)
